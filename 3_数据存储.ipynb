```python
# 查询
# 建立本地数据库连接
db = pymysql.connect(host='localhost', user=name, password=password, database='stockDataBase', charset='utf8')
cursor = db.cursor()

# 选择股票代码和股票简称列
cursor.execute('SELECT * FROM stockDataBase')
rows = cursor.fetchall()
```

```python
# 输出查询结果
for row in rows:
    print(row)
```

```python
# 生成 SQL 文件
with open('stockDataBase.sql', 'w', encoding='utf-8') as f:
    for row in rows:
        values = [str(i) if type(i) != str else f"'{i}'" for i in row]
        sql = f"INSERT INTO stockDataBase VALUES ({', '.join(values)});\n"
        f.write(sql)

# 关闭数据库连接
db.close()
```
